```python
def recursive_dict_print(d):
    ret_strings = []
    
    if type(d) == dict or type(d) == collections.OrderedDict:
        for k, v in d.items():
            x = recursive_dict_print(v)
            for l in x:
                ret_strings.append(k + '|' + l)
    elif type(d) == list:
        for elem in d:
            x = recursive_dict_print(elem)
            for l in x:
                ret_strings.append(l)
    else:
        ret_strings = [str(d)]
    
    return ret_strings
```
